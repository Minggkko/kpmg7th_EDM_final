```python
import os
import pandas as pd
from dotenv import load_dotenv
from supabase import create_client, Client

# 1. 환경 변수 로드
load_dotenv()
url: str = os.environ.get("SUPABASE_URL")
key: str = os.environ.get("SUPABASE_KEY")

# 2. Supabase 클라이언트 초기화
supabase: Client = create_client(url, key)

def process_and_upload():
    # 3. 데이터 가공 (예: 이전 SQLite에서 가져온 데이터라고 가정)
    data = [
        {"site_id": "Site A", "metric_name": "electricity", "unit": "MWh", "value": 1250.5, "v_status": 0},
        {"site_id": "Site B", "metric_name": "lng", "unit": "Nm3", "value": 88000.0, "v_status": 0}
    ]
    df = pd.DataFrame(data)

    # 4. 데이터 삽입 (테이블이 이미 존재해야 함)
    # 팁: 테이블 생성은 Supabase SQL Editor에서 미리 수행하는 것이 가장 안전합니다.
    try:
        response = supabase.table("standard_usage").insert(data).execute()
        print("✅ 데이터 삽입 성공:", response.data)
    except Exception as e:
        print("❌ 오류 발생:", e)

    # 5. 데이터 조회 및 필터링 (가공)
    # 예: v_status가 0인 데이터만 가져와서 확인
    query = supabase.table("standard_usage").select("*").eq("v_status", 0).execute()
    result_df = pd.DataFrame(query.data)
    
    print("📊 가공 대상 데이터:\n", result_df)

if __name__ == "__main__":
    process_and_upload()
```
